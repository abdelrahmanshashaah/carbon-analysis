```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
df = pd.read_csv('carbon_emission_dataset_with_Industry.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.columns
```

```python
cols = [
    'Total_Energy_Consumption_kWh',
    'Renewable_Energy_Consumption_kWh',
    'NonRenewable_Energy_Consumption_kWh',
    'Production_Output_Units',
    'Supply_Chain_Transport_km'
]

for col in cols:
    print(f"\n=== {col} ===")
    print(f"Mean:   {df[col].mean():.2f}")
    print(f"Median: {df[col].median():.2f}")
    print(f"Max:    {df[col].max():.2f}")
    print(f"Min:    {df[col].min():.2f}")
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Total_Energy_Consumption_kWh'], kde=True, bins=30)
plt.title('Distribution of Total Energy Consumption')
plt.xlabel('kWh')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df['Industry_Sectors'].value_counts().plot(kind='bar')
plt.title('Frequency of Industry Sectors')
plt.xlabel('Sector')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
daily_energy = df.groupby('Date')[['Total_Energy_Consumption_kWh',
                                    'Renewable_Energy_Consumption_kWh',
                                    'NonRenewable_Energy_Consumption_kWh']].sum()

plt.figure(figsize=(12, 6))
daily_energy.plot()
plt.title('Energy Consumption Over Time')
plt.ylabel('kWh')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['Date'], df['Carbon_Emission_tCO2e_TARGET'], alpha=0.5)
plt.title('Carbon Emissions Over Time')
plt.ylabel('tCO2e')
plt.show()
```

```python
# مقارنة زوجية بين مصادر الطاقة والانبعاثات
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Total Energy vs Emissions
axes[0].scatter(df['Total_Energy_Consumption_kWh'], df['Carbon_Emission_tCO2e_TARGET'], alpha=0.5)
axes[0].set_xlabel('Total Energy')
axes[0].set_ylabel('Carbon Emissions')
axes[0].set_title('Total Energy vs Emissions')
axes[1].scatter(df['Renewable_Energy_Consumption_kWh'], df['Carbon_Emission_tCO2e_TARGET'], alpha=0.5, color='green')
axes[1].set_xlabel('Renewable Energy')
axes[1].set_ylabel('Carbon Emissions')
axes[1].set_title('Renewable Energy vs Emissions')
axes[2].scatter(df['NonRenewable_Energy_Consumption_kWh'], df['Carbon_Emission_tCO2e_TARGET'], alpha=0.5, color='red')
axes[2].set_xlabel('NonRenewable Energy')
axes[2].set_ylabel('Carbon Emissions')
axes[2].set_title('NonRenewable Energy vs Emissions')

plt.tight_layout()
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
corr_matrix = df[numeric_cols].corr()
plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Industry_Sectors', y='Carbon_Emission_tCO2e_TARGET')
plt.title('Carbon Emissions by Industry Sector')
plt.xticks(rotation=45)
plt.show()
```

```python
selected_features = [
    'Carbon_Emission_tCO2e_TARGET',
    'Total_Energy_Consumption_kWh',
    'Renewable_Energy_Consumption_kWh',
    'Production_Output_Units',
    'Supply_Chain_Transport_km'
]

sns.pairplot(df[selected_features], diag_kind='kde')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
```

```python
daily_emissions = df.groupby('Date')['Carbon_Emission_tCO2e_TARGET'].sum()

plt.figure(figsize=(14, 6))
daily_emissions.plot()
plt.title('Carbon Emissions Over Time')
plt.ylabel('tCO2e')
plt.xlabel('Date')
plt.show()
```

```python
daily_emissions_ma7 = daily_emissions.rolling(window=7).mean()
daily_emissions_ma30 = daily_emissions.rolling(window=30).mean()

plt.figure(figsize=(14, 6))
plt.plot(daily_emissions, label='Daily', alpha=0.5)
plt.plot(daily_emissions_ma7, label='7-Day MA', color='orange')
plt.plot(daily_emissions_ma30, label='30-Day MA', color='red')
plt.title('Carbon Emissions with Moving Averages')
plt.ylabel('tCO2e')
plt.legend()
plt.show()
```

```python
rolling_var = daily_emissions.rolling(window=30).var()

plt.figure(figsize=(14, 6))
rolling_var.plot(color='purple')
plt.title('Rolling Variance (30-Day Window)')
plt.ylabel('Variance')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
daily_emissions_filled = daily_emissions.asfreq('D').fillna(method='ffill')
decomposition = seasonal_decompose(daily_emissions_filled, model='additive', period=30)
fig = decomposition.plot()
fig.set_size_inches(14, 10)
plt.show()
```

```python

sources = [
    'Total_Energy_Consumption_kWh',
    'Renewable_Energy_Consumption_kWh',
    'NonRenewable_Energy_Consumption_kWh',
    'Production_Output_Units',
    'Supply_Chain_Transport_km'
]

variability = df[sources].std().sort_values(ascending=False)
print("Standard Deviation (Variability):")
print(variability)

# رسم
plt.figure(figsize=(10, 5))
variability.plot(kind='bar')
plt.title('Variability of Emission Sources')
plt.ylabel('Standard Deviation')
plt.xticks(rotation=45)
plt.show()
```

```python
from scipy import stats
sample = df['Carbon_Emission_tCO2e_TARGET'].sample(5000)
shapiro_stat, p_value = stats.shapiro(sample)
print(f"Shapiro-Wilk Test:")
print(f"Statistic: {shapiro_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Normal Distribution: {'Yes' if p_value > 0.05 else 'No'}")
plt.figure(figsize=(10, 5))
sns.histplot(df['Carbon_Emission_tCO2e_TARGET'], kde=True, bins=50)
plt.title('Distribution of Carbon Emissions')
plt.show()
```

```python
correlations = df[numeric_cols].corr()['Carbon_Emission_tCO2e_TARGET'].abs().sort_values(ascending=False)

print("Top correlations with Carbon Emissions:")
print(correlations.head(10))

plt.figure(figsize=(10, 5))
correlations.drop('Carbon_Emission_tCO2e_TARGET').head(10).plot(kind='barh')
plt.title('Top 10 Correlations with Carbon Emissions')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.plot(daily_emissions, alpha=0.3, label='Daily')
plt.plot(daily_emissions.rolling(60).mean(), color='red', linewidth=2, label='60-Day Trend')
plt.title('Trend in Carbon Emissions')
plt.legend()
plt.show()
first_month = daily_emissions.head(30).mean()
last_month = daily_emissions.tail(30).mean()
print(f"First month avg: {first_month:.2f}")
print(f"Last month avg: {last_month:.2f}")
print(f"Trend: {'Increasing' if last_month > first_month else 'Decreasing'}")
```

```python
df['Month'] = df['Date'].dt.month
monthly_avg = df.groupby('Month')['Carbon_Emission_tCO2e_TARGET'].mean()

plt.figure(figsize=(10, 5))
monthly_avg.plot(kind='bar', color='skyblue')
plt.title('Average Carbon Emissions by Month')
plt.xlabel('Month')
plt.ylabel('tCO2e')
plt.xticks(rotation=0)
plt.show()
print("Monthly averages:")
print(monthly_avg)
```
